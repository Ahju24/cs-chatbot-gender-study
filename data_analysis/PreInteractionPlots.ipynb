# Results of Pre-interaction Questions
This script contains the visualizations for the pre-interaction question response distributions.

### Content:
* Prior Programming Experience
* Frequency of Chatbot Usage

```python
import matplotlib.pyplot as plt
import pandas as pd

color = ['lightblue', 'lightpink', 'lightcoral', 'lightgreen']

# Prior programming experience
height = [7, 64, 0, 23]
bars = ('No experience', 'Basic knowledge', 'Intermediate', 'Advanced')
df = pd.DataFrame({'height': height,
                   'bars': bars})

# Create bars
fig, ax = plt.subplots()
ax.bar(df['bars'], df['height'],
       color=color)

# Show graph
plt.show()
```

```python
color = ['lightblue', 'lightpink', 'lightgreen', 'lightcoral', 'plum']

# Frequency of chatbot usage
height = [49, 31, 11, 2, 1]
bars = ('Daily', 'Frequently', 'Occasionally', 'Rarely', 'Never')
df = pd.DataFrame({'height': height,
                   'bars': bars})

# Create bars
fig, ax = plt.subplots()
ax.bar(df['bars'], df['height'],
       color=color)

# Show graph
plt.show()
```
